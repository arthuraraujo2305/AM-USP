# Importando bibliotecas

```python
# Básicos
import pandas as pd
import numpy as np
```

## Carregando os dados "Housing.csv" em um DataFrame.

```python
DATA_PATH = "./Housing.csv"
```

```python
df = pd.read_csv(DATA_PATH)
df.head()
```

## 1. Pré-processamento

Realize o pré-processamento dos dados acima considerando os seguintes etapas.


1.   Valide que não existem valores nulos dentre os dados carregados.
2.   Separe os atributos da base de dados em **numéricos** e **categóricos**.
3.   Descarte os atributos categóricos.
4.   Isole a variável a ser predita "price".
4.   Separe os dados em treino e teste, considerando 20% dos dados para teste e usando a seed 2026.
5.   Finalize padronizando os dados divididos.



```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Verificando valores nulos
print("Valores nulos por coluna:")
print(df.isnull().sum())
assert df.isnull().sum().sum() == 0, "Existem valores nulos!" #Desse jeito se existirem valores nulos, o código para 
                                                              #e mostra a mensagem de erro. Se não existirem, continua e mostra o print
print("\nNenhum valor nulo encontrado.\n")

#Transformando algumas colunas categóricas em numéricas
colunas_transformar = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']
df[colunas_transformar] = df[colunas_transformar].replace({'yes': 1, 'no': 0})

# Separar atributos numéricos e categóricos
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = df.select_dtypes(exclude=[np.number]).columns.tolist()
print(f"Atributos numéricos: {numeric_cols}")
print(f"Atributos categóricos: {categorical_cols}\n")

# Descartando atributos categóricos
df_numeric = df[numeric_cols].copy()

# Isolar a variável a ser predita "price"
y = df_numeric["price"]
X = df_numeric.drop(columns=["price"])

print(f"Features (X): {X.columns.tolist()}")
print(f"Target (y): price\n")

# 5. Separando os dados
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2026)
print(f"Treino: {X_train.shape[0]} amostras")
print(f"Teste:  {X_test.shape[0]} amostras\n")

# Padronizando
scaler = StandardScaler()
X_train = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns, index=X_train.index)
X_test = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns, index=X_test.index)
```

## 2. Avaliação de uma abordagem "ingênua"

 Leia a documentação do [*DummyRegressor*](https://scikit-learn.org/stable/modules/generated/sklearn.dummy.DummyRegressor.html) e o treine nos dados processados usando seus parâmetros padrões. Avalie a performance do modelo usando a métrica MSE.

```python
from sklearn.dummy import DummyRegressor
from sklearn.metrics import mean_squared_error
```

```python
#Aplicando dummy regressor
dummy = DummyRegressor(strategy="mean", constant=None, quantile=None)
dummy.fit(X_train, y_train)
y_pred_dummy = dummy.predict(X_test)
mse_dummy = mean_squared_error(y_test, y_pred_dummy)
print(f"Mean Squared Error do Dummy Regressor: {mse_dummy:.2f}")
```
